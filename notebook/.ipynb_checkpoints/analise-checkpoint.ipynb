## Analise dos dados

Analise feita com ajuste polinomial de $2^{nd}$ ordem.

Assumindo que os dados obedecem a equação: $f(x) = a x^2 + b x + c$

O estimados de minimos quadrados é:

$ p = (A^TA)^{-1}A^Td^o $

$ A_{i,1} = x_i^2 $

$ A_{i,2} = x_i $

$ A_{i,3} = 1 $

```python
def ajuste(x, y):
    A = np.transpose([x**2, x, np.ones_like(x)])
    a, b, c = np.linalg.solve(np.dot(A.T, A), np.dot(A.T, y))
    return a, b, c

def predito(a, b, c, x):
    return a*x**2 + b*x + c
```

```python
a, b, c = ajuste(x, y)
yp = predito(a, b, c, x)

print "Coeficientes estimados:", a, b, c

plt.plot(x, y, '.k', x, yp, '-r')
plt.savefig('ajuste.png')
```
